```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import time

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 10000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 2)
```

```python
df_weather = pd.read_csv('./weather_combined_november.csv', low_memory=False)
```

```python
df_weather['UTC_DATE'] = pd.to_datetime(df_weather['UTC_DATE'], utc=True)
df_weather['LOCAL_DATE'] = df_weather['UTC_DATE'].dt.tz_convert('America/New_York')
df_weather['LOCAL_DATE'] = df_weather['LOCAL_DATE'].dt.strftime('%Y-%m-%d')
```

```python
df_filter = df_weather[df_weather['LOCAL_DATE'] != '2020-11-15'] # Ontario Wind

df_year = df_filter.groupby(['LOCAL_YEAR'])["WIND_SPEED"].mean().reset_index()
display(df_year)
```

```python
# df_daily = df_weather.groupby(['LOCAL_DATE'])["WIND_SPEED"].agg(['mean', 'min', 'max', 'std', 'var']).reset_index()
df_station = df_filter.groupby(['LOCAL_HOUR','STATION_NAME'])["WIND_SPEED"].agg([
    'mean', 'min', 'max', 'std', 'var']).reset_index()

# display(df_daily)
display(df_station)
```

```python
# df_daily = df_weather.groupby(['LOCAL_DATE'])["WIND_SPEED"].agg(['mean', 'min', 'max', 'std', 'var']).reset_index()
df_hour = df_filter.groupby(['LOCAL_DATE', 'LOCAL_HOUR'])["WIND_SPEED"].agg([
    'mean', 'min', 'max', 'std', 'var']).reset_index()

# display(df_daily)
display(df_hour)
```
